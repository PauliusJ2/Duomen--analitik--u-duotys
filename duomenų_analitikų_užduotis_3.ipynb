## Naudojamos bibliotekos:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grupes = [
    {"pavadinimas": "1 pensininkas\n48 kv. m butas", "pajamos_max": 2000, "pajamos_step": 50, "saskaita_max": 200, "saskaita_step": 5, "nariu_skaicius": 1, "plotas": 48, "max_kompensacija": 100},
    {"pavadinimas": "1 pensininkas\n120 kv. m butas", "pajamos_max": 2000, "pajamos_step": 50, "saskaita_max": 200, "saskaita_step": 5, "nariu_skaicius": 1, "plotas": 120, "max_kompensacija": 50},
    {"pavadinimas": "2 dirbantys suaugę\n76 kv. m butas", "pajamos_max": 5000, "pajamos_step": 50, "saskaita_max": 500, "saskaita_step": 10, "nariu_skaicius": 2, "plotas": 76, "max_kompensacija": 70},
    {"pavadinimas": "1 dirbantis su 3 vaikais\n76 kv. m butas", "pajamos_max": 5000, "pajamos_step": 100, "saskaita_max": 500, "saskaita_step": 10, "nariu_skaicius": 4, "plotas": 76, "max_kompensacija": 100},
]
# Valstybės remiamų pajamų (VRP) dydis
VRP = 176 
PLOTO_NORMATYVAS = 50
SEIMOS_NORMATYVAS = {"pirmas": 38, "kiekvienas": 12}

def gauti_normatyvini_plota(nariu_skaicius, buto_plotas):
    if nariu_skaicius == 1:
        plotas = PLOTO_NORMATYVAS
    else:
        plotas = SEIMOS_NORMATYVAS["pirmas"] + (nariu_skaicius - 1) * SEIMOS_NORMATYVAS["kiekvienas"]
    return min(plotas, buto_plotas)

def pajamu_korekcija(nariu_skaicius):
    return 2 * VRP * nariu_skaicius if nariu_skaicius > 1 else 3 * VRP
    # Vienam asmeniui taikomas 3 x VRP, šeimai - 2 x VRP už kiekvieną narį

fig, axs = plt.subplots(2, 2, figsize=(16, 14))

for i, grupe in enumerate(grupes):
    pajamos = np.arange(0, grupe["pajamos_max"] + grupe["pajamos_step"], grupe["pajamos_step"])
    sildymo_saskaita = np.arange(0, grupe["saskaita_max"] + grupe["saskaita_step"], grupe["saskaita_step"])
    pajamos_grid, saskaitos_grid = np.meshgrid(pajamos, sildymo_saskaita)

    normatyvinis_plotas = gauti_normatyvini_plota(grupe["nariu_skaicius"], grupe["plotas"])
    efektyvi_saskaita = saskaitos_grid * (normatyvinis_plotas / grupe["plotas"])

    rmp_atimama = pajamu_korekcija(grupe["nariu_skaicius"])
    
    # Apskaičiuojama kompensuojama dalis
    kompensuojama = np.maximum(0, efektyvi_saskaita - 0.1 * (pajamos_grid - rmp_atimama))
    kompensacijos_procentai = (kompensuojama / (efektyvi_saskaita + 1)) * 100
    kompensacijos_procentai = np.minimum(kompensacijos_procentai, grupe["max_kompensacija"])

    ax = axs[i // 2, i % 2]
    c = ax.contourf(
        pajamos_grid,
        saskaitos_grid,
        kompensacijos_procentai,
        levels=np.arange(0, 110, 10),
        cmap="YlOrBr",
        extend="both",
    )
    # Grafiko pavadinimas ir ašių aprašymas
    ax.set_title(grupe["pavadinimas"], fontsize=16, pad=10)
    ax.set_xlabel("Suminės pajamos į rankas (EUR)", fontsize=12)
    ax.set_ylabel("Šildymo sąskaita (EUR)", fontsize=12)
    ax.grid(True, alpha=0.5, linestyle="--")

    cbar = fig.colorbar(c, ax=ax, orientation="vertical")
    cbar.set_label("Valstybės kompensuojama sąskaitos dalis (%)", fontsize=10)
    cbar.ax.tick_params(labelsize=10)

# Pakankami tarpai tarp grafikų, bendras pavadinimas visom lentelėms
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle("Būsto šildymo išlaidų kompensacijų skaičiavimas", fontsize=18, y=0.98)
plt.show()
```
